```python
import pandas as pd
import numpy as np
```

# Melt

```python
df = pd.DataFrame({'Stock': ['Goog','Yahoo','Mcsd'],
                   '2019-1-1': [1,2,3],
                   '2019-1-2': [1,2,3],
                   '2019-1-3': [1,2,3]})
display(df)
display(pd.melt(df, id_vars=['Stock']))
```

# Pivot Table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "large", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
display(df)
display(pd.pivot_table(df, index=['A', 'B'],
                    columns=['C'],values=['D'], aggfunc='mean'))

```

```python
display(df)
display(pd.pivot_table(df, index=['A', 'B'], columns=['C'],aggfunc='sum'))
display(pd.pivot_table(df, index=['A', 'B','C'], aggfunc='sum'))
```

```python
df1=pd.pivot_table(df, index=['A', 'B'],
                    columns=['C'], aggfunc='count')
display(df1)
df1.columns=["_".join(x) for x in df1.columns.ravel()]

display(df1)
```
